```python
import requests
from bs4 import BeautifulSoup
import pprint
import pandas as pd

url = 'https://www.glassdoor.com/Job/houston-python-jobs-SRCH_IL.0,7_IC1140171_KO8,14.htm?seniorityType=entrylevel'
headers = {'User-agent':'Mozilla/5.0'}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'html.parser')

jobcards = soup.find_all('li', 'react-job-listing')

def get_record(card):
    company = card.find('div', 'd-flex justify-content-between align-items-start').a.text
    job_title= card.find('div','d-flex justify-content-between align-items-start').find_next('a').find_next('a').text
    job_url = ("https://glassdoor.com" + card.find('div', 'd-flex justify-content-between align-items-start').a.get('href'))

    try:
        salary_estimate = card.find('div', 'css-1buaf54 pr-xxsm').find_next('span').getText()
    except AttributeError:
        salary_estimate = ''
        
    company_rating = card.find('div', 'd-flex flex-column job-search-key-1pzmdmc e1rrn5ka1').find_next('span').find_next('span').text
    post_date = card.find('div','d-flex align-items-end pl-std css-17n8uzw').text
    
    record = (company, job_title, salary_estimate, company_rating, job_url)
    return record


records=[]

for card in jobcards:
    record = get_record(card)
    records.append(record)
    
pprint.pprint(records)
```
